joinmfunctio - length of the value should be same 

```python
import pandas as pd 
dic1=pd.DataFrame({"a":[3,5,4,6,4],"b":[5,7,85,4,74]})
dic2=pd.DataFrame({"c":[3,5,4,6,4],"d":[5,7,85,4,3]})
dic1.join(dic2)
```

#marge-- flter out comman data and join them

```python

dic3=pd.DataFrame({"s":[31,5,4,6,4,8],"h":[9,5,73,85,4,74]})
dic4=pd.DataFrame({"s":[5,4,6,8],"r":[5,76,85,4]})
pd.merge(dic3,dic4,how="outer",indicator=True)
```
